```python
oil_prices = pd.read_csv('/content/BrentOilPrices.csv')
oil_prices['Date'] = pd.to_datetime(oil_prices['Date']).dt.strftime('%Y-%m-%d')
oil_prices.rename(columns={'Price':'Oil_price'}, inplace=True)

oil_prices.head(50)
```

```python
oil_prices['Date'] = pd.to_datetime(oil_prices['Date'])

oil_prices.info()
```

```python
data = bitcoin.merge(oil_prices, how='inner', on='Date')

data.head()
```

```python
data['Diff_high_low'] = data['High'] - data['Low']
data['Daily_gain'] = data['Close'] - data['Open']

data.head()
```

```python
data.describe()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Close'], marker='o')
plt.title('Serie de Tiempo de Precios de Cierre de Bitcoin')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Oil_price'], marker='o')
plt.title('Serie de Tiempo de Precios de Crudo Brent')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre')
plt.grid(True)
plt.show()
```

```python
corr = data.corr()
sns.heatmap(corr, annot=True)
```
